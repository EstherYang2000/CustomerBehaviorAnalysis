#### Phase 2 加權平均購買天數

要算出每個會員的 purchase date list 並做加權

```python
def days_between(d1, d2):
    d1 = datetime.strptime(d1[:10], "%Y-%m-%d")
    d2 = datetime.strptime(d2[:10], "%Y-%m-%d")
    if abs((d2 - d1).days) == 0: return 1
    return abs((d2 - d1).days) 

def get_cai(MemberId, RegisterDateTime):
    cur_member = MemberId
    cur_member_register_date = RegisterDateTime
    cur_member_history_data = Order[Order['MemberId'] == cur_member]
    purchase_dates = cur_member_history_data['OrderDateTime'].tolist()
    purchase_dates.sort()
    purchase_dates.insert(0, cur_member_register_date)
    purchase_dates.append(timestamp_today)
    purchase_count = len(purchase_dates) - 1 # 減掉一開始第一個是註冊會員的日期
    total_days = days_between(purchase_dates[-1], purchase_dates[0])
    average_purchase_days = total_days / purchase_count
    period_list = []
    for i in range(len(purchase_dates) - 1):
        period_list.append(days_between(purchase_dates[i], purchase_dates[i+1]))
    weighted_score = 0
    weight_sum = 0
    for index, period in enumerate(period_list):
        weighted_score += (index + 1) * period
        weight_sum += (index + 1)
    weighted_average_purchase_days = weighted_score / (weight_sum)
    cai = (average_purchase_days - weighted_average_purchase_days) / average_purchase_days
    return cai
```

### 資料前處理

```python
def ProcessingData(behavior): 
    behavior = behavior.drop(['Tunnel', 'Device',  'HitTime','FullvisitorId','DeviceId',
           'Language', 'CountryAliasCode', 'Version', 'UTMSource', 'UTMMedium',
           'UTMName',  'RegisterTunnel', 'CategoryId', 'SalePageId',
           'UnitPrice', 'Qty', 'TotalSalesAmount', 'CurrencyCode',
           'TradesGroupCode', 'SearchTerm', 'ContentType', 'ContentName',
           'ContentId', 'PageType'],axis=1)
    print('drop columns')
    behavior = behavior.drop(behavior[behavior['MemberId'].isnull()].index,axis=0).reset_index(drop=True)
    print('drop null MemberId')
    groupbehavior = behavior.groupby(['MemberId','Behavior'],as_index=False)
    groupbehavior = groupbehavior.count()
    print('groupbehavior')
    groupbehavior.rename({'EventTime':'countime'}, axis=1)
    print('rename')
    score = {'register':3,'viewproduct':8,'search':8,'add':9,'checkout':10,'purchase':10,'selectcontent':7,'viewmainpage':8,
             'viewcategory':8,'viewactivity':8,'viewcustompage':7,'viewalbumdetail':8,'viewvideodetail':7,'viewarticledetail':7,
             'viewecoupondetail':9,'viewpromotiondetail':9,'viewcoupondetail':9}
    groupbehavior['Behavior'] = groupbehavior['Behavior'].map(score)
    print('map')
    groupbehavior['TotalBehaviorScore'] = groupbehavior['Behavior'] * groupbehavior['EventTime']
    print('BehaviorScore')
    groupbehavior = groupbehavior.groupby(['MemberId'])['TotalBehaviorScore','EventTime'].sum().reset_index()
    print('sum')
    groupbehavior['TotalBehaviorScore'] = groupbehavior['TotalBehaviorScore']/groupbehavior['EventTime']
    print('TotalBehaviorScore')
    groupbehavior = groupbehavior.drop(['EventTime'],axis=1)
    print('dropEventTime')
    return groupbehavior
```
